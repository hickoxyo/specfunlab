### Error Propagation

```python
# This lesson and examples were inspired by the work of Earl Blodgett (UW Riverfalls)

import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

However in the real world we generally don't have knowlege of the actual error in our measurement... if we knew exactly what we did wrong we would try to correct it.  Instead we have an estimation of the uncertainity of our measurement that we usually derive from statistics.

I say usually because the process of estimating uncertainty depends on what you are measuring and that doesn't always have clear guidelines... or straightforward statistics.  However, what we do have is a formula/system/recipe that has been developed over the years which gives us a best estimation of our uncertainity in measurement.

Below we are going to compare two methods of error propagation.  The formal method I just described above and a statistical method using Monte Carlo estimations.  As a first example we will consider the uncertanity it the sum of two measurements.

```python
N = 100000

# Here we are defining the measured values we may have taken.
mu_x1 = 5.00
sigma_x1 = .20
mu_x2 = 6.50
sigma_x2 = .10

# Now, if we make the assumption that our measurements are normally distributed about the average value we can create a distribution.
x1 = np.random.normal(mu_x1,sigma_x1, N)
x2 = np.random.normal(mu_x2,sigma_x2, N)

# This is our actual function as well as the mean and standard deviation using our simulated distribution of 100000 measurements.
my_function = x2 + x1
my_average = np.mean(my_function)
my_error = np.std(my_function)

# If we were in the lab we would estimate our error using the traditional method for error propagation.
trad_average = mu_x2 + mu_x1
trad_error = math.sqrt(sigma_x2**2 + sigma_x1**2)

# We can compare our simulated error (Monte Carlo method) to our traditional calculation.
print("Monte Carlo Result: ",my_average, "  ", my_error)
print("Traditional Result: ",trad_average, "  ", trad_error)


count, bins, ignored = plt.hist(my_function, 30, density=True)
# create 30 points of a gaussian function based on the average and error
overlay = 1/(my_error * np.sqrt(2 * np.pi)) * np.exp( - (bins - my_average)**2 / (2 * my_error**2) )
# add that gaussian curve to the graph
plt.plot(bins,overlay)
plt.show()
```

As a second example we could consider the error in the product of two measured quantities.

```python
N = 100000

mu_x1 = 5.00
sigma_x1 = .20
mu_x2 = 6.50
sigma_x2 = .10

x1 = np.random.normal(mu_x1,sigma_x1, N)
x2 = np.random.normal(mu_x2,sigma_x2, N)

my_function = x2 / x1

my_average = np.mean(my_function)
my_error = np.std(my_function)

trad_average = mu_x2 / mu_x1
trad_error = trad_average * math.sqrt((sigma_x2/mu_x2)**2 + (sigma_x1/mu_x1)**2)

print("Monte Carlo Result: ",my_average, "  ", my_error)
print("Traditional Result: ",trad_average, "  ", trad_error)

count, bins, ignored = plt.hist(my_function, 30, density=True)
# create 30 points of a gaussian function based on the average and error
overlay = 1/(my_error * np.sqrt(2 * np.pi)) * np.exp( - (bins - my_average)**2 / (2 * my_error**2) )
# add that gaussian curve to the graph
plt.plot(bins,overlay)
plt.show()
```

What about putting this all together and considering a simple equation from general physics which contains both products and sums?

v=v0+at
where v0 = -2.40  0.20 m/s,a = 1.55  0.15 m/s2 and t = 12.0  1.0 s.

```python
N = 1000000

mu_v0 = -2.4
sigma_v0 = .2
mu_a = 1.55
sigma_a = .15
mu_t = 12
sigma_t = 1

v0 = np.random.normal(mu_v0, sigma_v0,N)
a = np.random.normal(mu_a, sigma_a, N)
t = np.random.normal(mu_t, sigma_t, N)

my_function = v0 + a * t

my_average = np.mean(my_function)
my_error = np.std(my_function)

trad_average = mu_v0 + mu_a * mu_t
trad_error = math.sqrt((sigma_v0)**2 + (sigma_a * mu_t)**2 + (sigma_t * mu_a)**2)

print("Monte Carlo Result: ",my_average, "  ", my_error)
print("Traditional Result: ",trad_average, "  ", trad_error)

count, bins, ignored = plt.hist(my_function, 30, density=True)
# create 30 points of a gaussian function based on the average and error
overlay = 1/(my_error * np.sqrt(2 * np.pi)) * np.exp( - (bins - my_average)**2 / (2 * my_error**2) )
# add that gaussian curve to the graph
plt.plot(bins,overlay)
plt.show()
```

Finally we can consider the situation where our function is more complicated and the assumption that it is linear near the measurement is not as easily verified.  We will consider the function 1/r^4.

```python
N = 1000000

mu_r = 5.00
sigma_r = .50

f = np.random.normal(mu_r, sigma_r, N)

my_function = pow(f,-4)

my_average = np.mean(my_function)
my_error = np.std(my_function)

trad_average = pow(mu_r,-4)
trad_error = 4*trad_average * sigma_r/mu_r

print("Monte Carlo Result: ",my_average, "  ", my_error)
print("Traditional Result: ",trad_average, "  ", trad_error)

center_slope = 4 * pow(mu_r, -3)
low_slope = 4 * pow(mu_r - sigma_r, -3)
high_slope = 4 * pow(mu_r + sigma_r, -3)

percent_change = 50 * abs((high_slope - low_slope)/center_slope)
print("the slope", center_slope," varies from ", low_slope, " to ", high_slope)
print("slope varies by about ", round(percent_change),  "%")

count, bins, ignored = plt.hist(my_function, 30, density=True)
# create 30 points of a gaussian function based on the average and error
overlay = 1/(my_error * np.sqrt(2 * np.pi)) * np.exp( - (bins - my_average)**2 / (2 * my_error**2) )
# add that gaussian curve to the graph
plt.plot(bins,overlay)
plt.show()
```
