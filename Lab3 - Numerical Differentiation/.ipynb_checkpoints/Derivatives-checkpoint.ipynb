ONE: The first is that you are analyzing data that you have collected AND maybe already processed to some degree.

```python
col_hdrs = ['1','volts','3','4','5','6','amperes']
IV = pd.read_excel(r'IV_LED_Data.xlsx', names = col_hdrs)
IV.head()
```

```python
i = np.array(IV.amperes)
v = np.array(IV.volts)
print('Number of readings: ', len(i))
plt.plot(v,i)
plt.title('Raw IV data')
plt.xlabel('volts')
_ = plt.ylabel('amps')
```
