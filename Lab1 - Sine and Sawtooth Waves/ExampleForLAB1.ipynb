## Example
### Plot a sine function

#### Write a program to plot a sine wave. 
#### Use sampling rate of 800 (samples/Sec), duration of 2 Sec., and frequency 6 Hz.
You may use other values to experiment

```python
# Include all imports in the first cell upfront
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants in this cell
# samples per second
s = 800
# frequency
f = 1
# duration
d = (-np.pi,np.pi)

# note: the use of *d in the following statement
# create time steps
t = np.linspace(*d, 2*s)
```

```python
# Calculate sin(theta)
y = np.sin(2*np.pi*f*t)
```

```python
# Plot y at times t
p = plt.plot(t,y)

# Note: always label your plot
plt.title("Sine Wave")
plt.xlabel("Sec")
amp = plt.ylabel("Amplitude")
```

#### Add another sine wave

```python
y1 = np.sin(2*np.pi*3*f*t)/3
```

```python
p = plt.plot(t, y+y1)
```
