#### Start by writing a program to plot a sine wave. Use duration from -$\pi$ to  $\pi$ and a certain frequency. Add sine waves with different frequencies until your plot resembles either a square wave or a saw-tooth wave.

```python
# Include all imports in this cell
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants in this cell
#samples rate
s=800

#amplitude
a=3.5

#frequency
f=1

#duration
d=(-np.pi,np.pi)
```

```python
# Calculate time values (t) and sin(t)

#calculate t
t=np.linspace(*d,2*s)

#calculate sin
y=a*np.sin(2*np.pi*f*t)
```

```python
# Put your plot statements here

p=plt.plot(t,y)

#define a second sine wave
y2=np.sin(2*np.pi*3*f*t)
p = plt.plot(t,y+y2)

plt.xlabel("seconds")
plt.ylabel("amplitude")
plt.title("sine wave")
```

```python
#define a second sine wave
y1=(a/1)*np.sin(2*np.pi*1*f*t)
y2=(a/3)*np.sin(2*np.pi*3*f*t)
y3=(a/5)*np.sin(2*np.pi*5*f*t)
y4=(a/7)*np.sin(2*np.pi*7*f*t)
y5=(a/9)*np.sin(2*np.pi*9*f*t)
y6=(a/11)*np.sin(2*np.pi*11*f*t)
y7=(a/13)*np.sin(2*np.pi*13*f*t)
y8=(a/15)*np.sin(2*np.pi*15*f*t)

#plot square wave
p = plt.plot(t,y+y2+y3+y4+y5+y6+y7+y8)
plt.xlabel("seconds")
plt.ylabel("amplitude")
plt.title("square wave")
```
