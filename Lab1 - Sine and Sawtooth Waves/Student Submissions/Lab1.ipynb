#### Start by writing a program to plot a sine wave. Use duration from -$\pi$ to  $\pi$ and a certain frequency. Add sine waves with different frequencies until your plot resembles either a square wave or a saw-tooth wave.

```python
# Include all imports in this cell
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants in this cell
#sample rate
s=800
#Sample Amplitude
a=3.5
#frequency in hertz
f=5
#duration (start and endpoints)
d= (-1, 1)
```

```python
# Calculate time values (t) and sin(t)
#calculate t
t= np.linspace(-1, 1 , 2*s)
#calculate sin
y = a*np.sin(2*np.pi*f*t)
y2= (a/2)*np.sin(2*np.pi*f*2*t)
y3= (a/3)*np.sin(2*np.pi*f*3*t)
y4= (a/4)*np.sin(2*np.pi*f*4*t)
y5= (a/5)*np.sin(2*np.pi*f*5*t)
y6= (a/6)*np.sin(2*np.pi*f*6*t)
y7= (a/7)*np.sin(2*np.pi*f*7*t)
y8= (a/8)*np.sin(2*np.pi*f*8*t)
y9= (a/9)*np.sin(2*np.pi*f*9*t)
y10= (a/10)*np.sin(2*np.pi*f*10*t)
```

```python
# Put your plot statements here
p= plt.plot(t, y+y2+y3+y4+y5+y6+y7+y8+y9+y10)
plt.xlabel('seconds')
plt.ylabel('amplitude')
plt.title('sine wave')
```
