#### Start by writing a program to plot a sine wave. Use duration from -$\pi$ to  $\pi$ and a certain frequency. Add sine waves with different frequencies until your plot resembles either a square wave or a saw-tooth wave.

```python
# Include all imports in this cell

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants in this cell

s = 800            # Sample rate
a = 5              # Amplitude
f = 1              # Frequency
d = (-np.pi,np.pi) # Duration
```

```python
# Calculate time values (t) and sin(t)

t = np.linspace(*d,2*s) # Time values

y = a*np.sin(2*np.pi*f*t)         # Sin values
```

```python
# Put your plot statements here

p = plt.plot(t,y)       # Plot y as function of t
plt.xlabel("Seconds")   # Set x label
plt.ylabel("Amplitude") # Set y label
plt.title("Sine Wave")  # Set graph title
```

```python
# Additional sine functions

y2 = 3 * np.sin(2*np.pi*3*f*t)

p2 = plt.plot(t,y+y2)
```

```python
saw1 = (a/1)*np.sin(1*(2*np.pi)*f*t)
saw2 = -(a/2)*np.sin(2*(2*np.pi)*f*t)
saw3 = (a/3)*np.sin(3*(2*np.pi)*f*t)
saw4 = -(a/4)*np.sin(4*(2*np.pi)*f*t)
saw5 = (a/5)*np.sin(5*(2*np.pi)*f*t)
saw6 = -(a/6)*np.sin(6*(2*np.pi)*f*t)
saw7 = (a/7)*np.sin(7*(2*np.pi)*f*t)
saw8 = -(a/8)*np.sin(8*(2*np.pi)*f*t)
saw9 = (a/9)*np.sin(9*(2*np.pi)*f*t)
saw10 = -(a/10)*np.sin(10*(2*np.pi)*f*t)

plt.plot(t,saw1+saw2+saw3+saw4+saw5+saw6+saw7+saw8+saw9+saw10)
```
