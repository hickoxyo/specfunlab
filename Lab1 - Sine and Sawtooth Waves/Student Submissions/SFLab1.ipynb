```python
# Include all imports in this cell

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants in this cell

#sample rate
s=1000
#amplitude
a=4
#frequency
f=1
#duration
d=(-np.pi,np.pi)
```

```python
# Calculate time values (t) and sin(t)

#calculate t
t=np.linspace(-np.pi,np.pi,2*s)
#calculate sin
y=a*np.sin(2*np.pi*f*t)
```

```python
# Put your plot statements here

p=plt.plot(t,y)
plt.xlabel("seconds")
plt.ylabel("amplitude")
plt.title("Sign of a Sine")
```

```python
# Define a second sine wave

y2=a/2*np.sin(2*np.pi*2*f*t)
y3=a/3*np.sin(2*np.pi*3*f*t)
y4=a/4*np.sin(2*np.pi*4*f*t)
y5=a/5*np.sin(2*np.pi*5*f*t)
y6=a/6*np.sin(2*np.pi*6*f*t)
y7=a/7*np.sin(2*np.pi*7*f*t)
y8=a/8*np.sin(2*np.pi*8*f*t)
y9=a/9*np.sin(2*np.pi*9*f*t)
y10=a/10*np.sin(2*np.pi*10*f*t)
y11=a/11*np.sin(2*np.pi*11*f*t)
y12=a/12*np.sin(2*np.pi*12*f*t)
y13=a/13*np.sin(2*np.pi*13*f*t)
y14=a/14*np.sin(2*np.pi*14*f*t)
y15=a/15*np.sin(2*np.pi*15*f*t)

yn = y+y2+y3+y4+y5+y6+y7+y8+y9+y10+y11+y12+y13+y14+y15 #store summation of frequencies as yn



p=plt.plot(t,yn)
```
