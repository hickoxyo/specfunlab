#### Start by writing a program to plot a sine wave. Use duration from -$\pi$ to  $\pi$ and a certain frequency. Add sine waves with different frequencies until your plot resembles either a square wave or a saw-tooth wave.

```python
# Include all imports in this cell
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants in this cell
#sample rate
s=800
#amplitude
a=3.5
#frequency
f=1
#duration
d=(-np.pi,np.pi)
```

```python
# Calculate time values (t) and sin(t)
#calculate t
t=np.linspace(*d,2*s)
#calculate sin
y=a*np.sin(2*np.pi*f*t)
```

```python
# Put your plot statements here
p=plt.plot(t,y)
plt.xlabel("seconds")
plt.ylabel("amplitude")
plt.title("sinewave")
```

```python
#define a second sine wave
y2=np.sin(1*np.pi*f*t)
p=plt.plot(t,y2)
```

```python
#define a third sine wave
y3=np.sin(3*np.pi*f*t)/3
p=plt.plot(t,y2+y3)
```

```python
#define a fourth sine wave
y4=np.sin(5*np.pi*f*t)/5
p=plt.plot(t,y2+y3+y4)
```

```python
#define a fifth sine wave
y5=np.sin(7*np.pi*f*t)/7
p=plt.plot(t,y2+y3+y4+y5)
```

```python
#define a sixth sine wave
y6=np.sin(9*np.pi*f*t)/9
p=plt.plot(t,y2+y3+y4+y5+y6)
```

```python
#define a seventh sine wave
y7=np.sin(11*np.pi*f*t)/11
p=plt.plot(t,y2+y3+y4+y5+y6+y7)
```
