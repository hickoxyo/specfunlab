### Part 1
Write a bit of python code to solve a simple radioactive decay problem.  A single step process with one decay rate.  Plot your Result.

We will use Radon-222 as an example.  The half-life of Mo-99 is 3.8 days.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from threading import Thread
from time import sleep, perf_counter
%matplotlib inline
```

```python
# Radon 222 half life - 3.8 Days

radon222HL = 3.8

# Decay rate constant
radon222RC = 0.693 / radon222HL

# Initial count
radon222Num = 10000
time = np.arange(0,3.8*8)

def dydt(y, t, r):
    return -r*y

decay = odeint(dydt, radon222Num, time, (radon222RC,))
plt.plot(time, decay[:,0])
```

### Part 2
Now consider what sort of error there is between your solution and the actual solution.  You know the actual solution to the differential quations is a simple exponential decay.

```python
r222EXP = radon222Num * np.exp(-time*radon222RC)
plt.plot(time, r222EXP, label = 'analytic')
plt.plot(time, decay[:,0], 'ro', label = 'numeric')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Count')
plt.title('Decay Two Ways')
plt.figure()
plt.plot(time, abs(r222EXP-decay[:,0]))
plt.title('Difference')
```

### Part 3
To begin the semester we tried to use random numbers to simulate a radioactive decay problem.  Re-do that problem using our modeling of differential equations.
- Used coupled ODE solution
 -  $ r' = -rcr*r $          ---------- (1)
 -  $ p' = -rcp*p + r' $     ---------- (2) 
 -  $ l1' = -rcl1*l1 + p' $  ---------- (3)
 -  $ b' = -rcb*b + l1' $    ---------- (4)
 -  $ l2' = -rcl2*r + b'$    ---------- (5)
- Shows decay rate of each element with total count of atoms in the system

```python
# Funtion to implement the coupled ODE slution
def decayFunc(y, t, l, f=True):
    """ Computes decay rates for the nuclear decay chain of Radon which is
        Radon
        -> Polonium
        -> Lead 214
        -> Bismuth
        -> Lead 210
        y is a vector of intial nuclear counts of the elements in order of the decay chain
        t is the time interval in days
        l is the vector of rate constants of the elements in order of the decay chain
        f = True if lead210 decays else it accumulates
        Returns a vector of derivatives of each element in order
    """
    # unpack counts and rate constants
    r,p,l1,b,l2 = y
    lm1,lm2,lm3,lm4,lm5 = l
    
    # Radon decay
    rn = -r*lm1
    
    # Polonium decay
    pn = -p*lm2 + r*lm1
    
    # Lead 214 decay
    l1n = -l1*lm3 + p*lm2
    
    # Bismuth decay
    bn = -b*lm4 + l1*lm3

    # Lead 210 decay
    if f:
        ln2 = -l2*lm5 + b*lm4
    else:
        ln2 = b*lm4
            
    return [rn,pn,l1n,bn,ln2]

# define element properties and counts
# Radon222- Half Life, Lambda, Count
HL1 = 3.8
RC1 = np.log(2) / HL1
CT1 = 1000000

# Polonium218- Half Life, Lambda, Count
HL2 = 3.1*30
RC2 = np.log(2) / HL2
CT2 = 0

# Lead214- Half Life, Lambda, Count
HL3 = 27*30
RC3 = np.log(2) / HL3
CT3 = 0

# Bismuth214- Half Life, Lambda, Count
HL4 = 20*30
RC4 = np.log(2) / HL4
CT4 = 0

# Lead210- Half Life, Lambda, Count
HL5 = 22*12*30
RC5 = np.log(2) / HL5
CT5 = 0
```

```python
# Set initial count vector and time steps
Initval = [CT1, CT2, CT3, CT4, CT5]

# Time range is set from 1 to a multiple of maximum half life in days
Time = np.arange(1,max(HL1,HL2,HL3,HL4,HL5)*20,30)

# Compute decay
rate = odeint(decayFunc, Initval, Time, ([RC1, RC2, RC3, RC4, RC5],))

# Show results
Time = np.log10(Time)
plt.figure(figsize=(8,5))
plt.plot(Time, rate[:,0], Time, rate[:,1], Time, rate[:,2], 
         Time, rate[:,3], Time, rate[:,4], Time, rate.sum(axis=1), '--')
plt.title('Nuclear Deacy Of 222Radon', fontsize=14)
plt.xlabel('Log10(Days)')
plt.ylabel('Atom Counts')
_=plt.legend(['Radon', 'Polonium', 'Lead214', 'Bismuth', 'Lead210', 'Totals'])
```

```python
# Compute decay as led210 accumulates

# reset time
# Time range is set from 1 to a multiple of maximum half life in days
Time = np.arange(1,max(HL1,HL2,HL3,HL4,HL5)*20,30)

# Compute decay
rate = odeint(decayFunc, Initval, Time, ([RC1, RC2, RC3, RC4, RC5], False))

# Show results
Time = np.log10(Time)
plt.figure(figsize=(8,5))
plt.plot(Time, rate[:,0], Time, rate[:,1], Time, rate[:,2], Time, 
         rate[:,3], Time, rate[:,4], Time, rate.sum(axis=1), '--')
plt.title('Nuclear Deacy Of 222Radon', fontsize=14)
plt.xlabel('Log10(Days)')
plt.ylabel('Atom Counts')
_=plt.legend(['Radon', 'Polonium', 'Lead214', 'Bismuth', 'Lead210', 'Totals'])
```
