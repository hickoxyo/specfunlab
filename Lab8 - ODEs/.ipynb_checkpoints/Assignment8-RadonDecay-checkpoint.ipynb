In order to solve, we'll need to use coupled ODEs:
-  $r' = -k_r r$           
 -  $ p' = -k_p p + r' $     
 -  $ l1' = -k_{l1} l1 + p' $   
 -  $ b' = -k_b b + l1' $     
 -  $ l2' = -k_{l2} l2 + b'$
